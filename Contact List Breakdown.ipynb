```python
cols = ['Contact_id', 'home_phone']
home_phone_df = df.loc[:, cols]
```

```python
home_phone_df.columns = ['Contact_id', 'Number']
home_phone_df['Phone_type'] = 'home'
home_phone_df['Area_code'] = home_phone_df['Number']
home_phone_df['Contact_id'] = df['Contact_id']
```

```python
for i in range(0, home_phone_df.shape[0]):
    #print(home_phone_df.iloc[i, 3])
    if home_phone_df.iloc[i, 3] == home_phone_df.iloc[i, 3]:
        areacode = home_phone_df.iloc[i, 3]
        areacode = str (areacode)[:3]
        home_phone_df.iloc[i, 3] = areacode
        ##print(home_phone_df.iloc[i, 3])
print(home_phone_df)
```

```python
home_phone_df.to_csv('HomePh.csv', index=False, encoding='utf-8')
```
